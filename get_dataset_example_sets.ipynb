```python
from google.colab import files, drive
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split

drive.mount('/content/drive')
```

```python
def stratified_subsets(df, label_col):
    remaining = df.copy()
    subset_seed_ids = np.empty((9, 15), dtype=df.index.dtype)

    for i in range(9):
        subset, remaining = train_test_split(
            remaining,
            train_size=15,
            stratify=remaining[label_col],
            random_state=43 + i
        )
        subset_seed_ids[i] = subset.index.to_numpy()
    return subset_seed_ids, remaining
```

```python
files.upload()
qnli_val = pd.read_csv('qnli_val.csv')

existing_ids_qnli = np.load('qnli_few_shot_example_idx.npy').reshape(1, -1)

df_remaining_qnli = qnli_val[~qnli_val.index.isin(existing_ids_qnli.flatten())]

subset_seed_ids_qnli, _ = stratified_subsets(df_remaining_qnli, label_col="label")

qnli_subset_ids = np.vstack([existing_ids_qnli, subset_seed_ids_qnli])

np.save("qnli_subset_seed_ids.npy", qnli_subset_ids)
```
